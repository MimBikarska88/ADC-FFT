```python
#the analog signal is a sinusoid

sinusoid = lambda amplitude, frequency, phase, t: amplitude * np.sin(2*np.pi*(frequency*t - phase))

def generate_periodic_signal(duration, amplitude, frequency, phase, sampling_frequency_1, sinusoid):
     #measuring the signal at evenly spaced time points
     num_samples = int(sampling_frequency_1 * duration)
     t = np.arange(num_samples) / sampling_frequency_1
     x = sinusoid(amplitude, frequency, phase, t)
     #x = amplitude * np.sin(2 * np.pi * (frequency * t - phase))
     return x, t

def sample_signal(t, x, duration, sampling_frequency_2):
     #sampling on the the signal with lower frequency
     num_samples_2 = int(sampling_frequency_2 * duration)
     t_discrete = np.arange(num_samples_2) / sampling_frequency_2
     x_discrete = np.interp(t_discrete, t, x)
     return x_discrete, t_discrete

def display_graph(x, t, x2, t2, label, x_lim = [], a_lim = []):
     plt.plot(t, x, color='blue', linewidth=2.0, linestyle='-')
     plt.xlim([x_lim[0], x_lim[1]])
     plt.ylim([-1.2*a_lim[0], 1.2*a_lim[1]])
     plt.xlabel('Time (seconds)')
     plt.ylabel('Amplitude')
     plt.title(label)
     plt.stem(t2, x2, linefmt='r', markerfmt='ro', basefmt='None')
     plt.grid()
     plt.show()
    
def hear_signal(x, sampling_frequency):
     ipd.display(ipd.Audio(x, rate = sampling_frequency))

def generate_composite_periodic_signal(amplitudes, frequencies, phases, sinusoid, duration, sampling_frequency_1):
    signal_matrix = np.array([amplitudes, frequencies, phases]).T
    x_total = None
    t = None
    print(duration)
    print(sampling_frequency_1)
    N = int(sampling_frequency_1 * duration)
    t = np.arange(N) / sampling_frequency_1
    for row in signal_matrix:
        amplitude, frequency, phase = row
        x  = sinusoid(amplitude, frequency, phase, t)
        if x_total is None:
                x_total = x
        else:
                x_total += x
    return x_total, t
    
def illustrate_periodic_analog_signal(duration, amplitude, frequency, phase, sampling_frequency, sampling_frequency_2, sinusoid):
    x, t = generate_periodic_signal(duration, amplitude, frequency, phase, sampling_frequency, sinusoid);
    x_discrete, t_discrete = sample_signal(t, x, duration, sampling_frequency_2)
    label = 'Sinusoid with frequency = %.1f, amp = %.1f, and phase = %.1f' % (frequency, amplitude, phase)
    display_graph(x, t, x_discrete, t_discrete, label, [0, duration], [amplitude, amplitude])
    hear_signal(x, sampling_frequency)


def illustrate_composite_analog_signal(amplitudes, frequencies, phases, sinusoid, duration, sampling_frequency_1, sampling_frequency_2):
    x, t = generate_composite_periodic_signal(amplitudes, frequencies, phases, sinusoid, duration, sampling_frequency_1)
    x_discrete, t_discrete = sample_signal(t, x, duration, sampling_frequency_2)
    label = (
    "Sinusoid of composite signal with "
    f"frequencies={[f'{f:.1f}' for f in frequencies]}, "
    f"amplitudes={[f'{a:.1f}' for a in amplitudes]}, "
    f"phases={[f'{p:.1f}' for p in phases]}"
    )
    display_graph(x, t, x_discrete, t_discrete, label, [0, 1], [max(amplitudes)*1.5, max(amplitudes)*1.5])
    hear_signal(x, sampling_frequency_1)

def hear_signal(x, sampling_frequency):
     ipd.display(ipd.Audio(x, rate = sampling_frequency))

def quantize_signal(x_discrete, bits):
    levels = 2**bits
    x_min = np.min(x_discrete)
    x_max = np.max(x_discrete)
    q = (xmax - xmin) / levels
    x_quantized = q * np.round(x_discrete / q)
    return x_quantized
    

    
illustrate_periodic_analog_signal(1,4,4,0,1000,100, sinusoid)
              
amplitudes  = [1,5,3]
frequencies = [1,4,1.6]
phases      = [0, 0, 0]
sampling_frequency_composite = 1000
sampling_frequency_composite_2 = 50
duration = 1

illustrate_composite_analog_signal(amplitudes, frequencies, phases, sinusoid, duration, sampling_frequency_composite, sampling_frequency_composite_2);

```
